# FM1 실습 1: 첫 번째 LLM API 호출

## 학습 목표
- Anthropic API를 설치하고 API 키를 안전하게 설정하는 방법을 배운다
- Claude 모델에 첫 번째 API 호출을 실행한다
- 프롬프트와 Temperature 파라미터의 역할을 이해한다

## 사전 준비
1. [console.anthropic.com](https://console.anthropic.com) 에서 API 키 발급 (최소 $5 크레딧 필요)
2. 아래 셀을 순서대로 실행하세요 (셀 클릭 후 `Shift+Enter`)

````python
# 1단계: anthropic 라이브러리 설치
# Google Colab에는 기본적으로 설치되어 있지 않으므로 먼저 설치합니다
!pip install anthropic -q
print("✅ 설치 완료!")
````

````python
# 2단계: API 키 안전하게 입력하기
# getpass를 사용하면 입력한 키가 화면에 표시되지 않습니다
# ⚠️ 절대 API 키를 코드에 직접 입력하지 마세요!
import getpass

api_key = getpass.getpass("🔑 Anthropic API 키를 입력하세요: ")
print("✅ API 키가 설정되었습니다 (보안을 위해 표시되지 않습니다)")
````

````python
# 3단계: Anthropic 클라이언트 생성
import anthropic

# 클라이언트 객체를 만듭니다 - 이후 모든 API 호출에 사용됩니다
client = anthropic.Anthropic(api_key=api_key)
print("✅ 클라이언트 준비 완료!")
````

## 🚀 첫 번째 API 호출

이제 Claude에게 처음으로 질문을 보내보겠습니다!

````python
# 4단계: 첫 번째 API 호출
# messages.create()로 Claude에게 메시지를 보냅니다

response = client.messages.create(
    model="claude-sonnet-4-6",       # 사용할 모델
    max_tokens=1024,                  # 응답의 최대 토큰 수
    messages=[
        {
            "role": "user",
            "content": "파이썬이란 무엇인지 초등학생도 이해할 수 있게 3줄로 설명해줘"
        }
    ]
)

# 응답에서 텍스트만 추출
answer = response.content[0].text
print("Claude의 답변:")
print(answer)
print(f"\n📊 사용 토큰: 입력 {response.usage.input_tokens} / 출력 {response.usage.output_tokens}")
````

## 💼 다양한 업무 프롬프트 활용

Claude는 다양한 업무에 활용할 수 있습니다. 몇 가지 예시를 실습해봅시다.

````python
# 업무 활용 예시 1: 이메일 작성 도우미
def ask_claude(prompt, max_tokens=512):
    """Claude에게 질문하고 답변을 반환하는 간단한 함수"""
    response = client.messages.create(
        model="claude-sonnet-4-6",
        max_tokens=max_tokens,
        messages=[{"role": "user", "content": prompt}]
    )
    return response.content[0].text

# 이메일 초안 작성
email_prompt = """
다음 상황에 맞는 비즈니스 이메일 초안을 작성해줘:
- 상황: 고객사 미팅 일정을 다음 주 화요일 오후 2시로 변경 요청
- 수신자: 김철수 부장님
- 발신자: 영업팀 박지영
- 톤: 정중하고 간결하게
"""

print("=" * 50)
print("📧 이메일 초안:")
print("=" * 50)
print(ask_claude(email_prompt))
````

````python
# 업무 활용 예시 2: 보고서 요약
report_text = """
2024년 3분기 영업실적 보고서 요약:
전체 매출은 전분기 대비 12% 증가한 45억원을 달성했습니다. 
신규 고객 획득이 38건으로 목표 30건을 초과 달성했으며,
기존 고객 재구매율은 68%로 전분기 대비 5%p 향상되었습니다.
그러나 영업비용이 8.5억원으로 예산 대비 15% 초과하여
순이익률은 전분기 21%에서 18%로 하락했습니다.
4분기 목표는 매출 50억원, 영업비용 8억원 이내 통제입니다.
"""

summary_prompt = f"""다음 보고서를 경영진을 위한 3줄 핵심 요약으로 만들어줘:\n{report_text}"""

print("=" * 50)
print("📊 경영진 요약 (3줄):")
print("=" * 50)
print(ask_claude(summary_prompt))
````

## 🌡️ Temperature 파라미터 실험

**Temperature**는 AI 출력의 창의성/무작위성을 조절합니다:
- `0.0` : 항상 같은 답변 (업무용 권장)
- `1.0` : 매번 다른, 더 창의적인 답변 (창작용)

````python
# Temperature 비교 실험
test_prompt = "신제품 마케팅 슬로건을 한 줄로 만들어줘. 제품: 스마트 물병 (물 마실 시간을 알려주는 기능)"

print("🔵 Temperature = 0.0 (일관된 답변)")
for i in range(3):
    response = client.messages.create(
        model="claude-sonnet-4-6",
        max_tokens=100,
        temperature=0.0,   # 결정적: 항상 같은 답변
        messages=[{"role": "user", "content": test_prompt}]
    )
    print(f"  시도 {i+1}: {response.content[0].text.strip()}")

print()
print("🔴 Temperature = 1.0 (다양한 답변)")
for i in range(3):
    response = client.messages.create(
        model="claude-sonnet-4-6",
        max_tokens=100,
        temperature=1.0,   # 창의적: 매번 다른 답변
        messages=[{"role": "user", "content": test_prompt}]
    )
    print(f"  시도 {i+1}: {response.content[0].text.strip()}")
````

## 📝 정리

| 개념 | 설명 |
|------|------|
| `client.messages.create()` | Claude에게 메시지를 보내는 핵심 함수 |
| `model` | 사용할 Claude 모델 (sonnet = 균형, haiku = 빠름, opus = 최강) |
| `max_tokens` | 응답 최대 길이 (1토큰 ≈ 0.75단어) |
| `temperature` | 창의성 조절 (0=일관성, 1=다양성) |
| `response.content[0].text` | 응답 텍스트 추출 방법 |
