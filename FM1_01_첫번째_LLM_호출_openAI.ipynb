## 💼 다양한 업무 프롬프트 활용

Claude는 다양한 업무에 활용할 수 있습니다. 몇 가지 예시를 실습해봅시다.

```python
# Claude와 GPT를 선택해서 사용할 수 있는 공통 함수
def ask_ai(prompt, model_type="claude", max_tokens=512):
    if model_type == "claude":
        response = claude_client.messages.create(
            model="claude-3-5-sonnet-20240620",
            max_tokens=max_tokens,
            messages=[{"role": "user", "content": prompt}]
        )
        return response.content[0].text

    elif model_type == "gpt":
        response = gpt_client.chat.completions.create(
            model="gpt-4o",
            max_tokens=max_tokens,
            messages=[{"role": "user", "content": prompt}]
        )
        return response.choices[0].message.content

# 이메일 초안 작성 예시 (GPT 사용)
email_prompt = """
다음 상황에 맞는 비즈니스 이메일 초안을 작성해줘:
- 상황: 고객사 미팅 일정을 다음 주 화요일 오후 2시로 변경 요청
- 수신자: 김철수 부장님
- 발신자: 영업팀 박지영
"""

print("📧 GPT가 작성한 이메일 초안:")
print(ask_ai(email_prompt, model_type="gpt"))
```

```python
# 업무 활용 예시 2: 보고서 요약
report_text = """
2024년 3분기 영업실적 보고서 요약:
전체 매출은 전분기 대비 12% 증가한 45억원을 달성했습니다.
신규 고객 획득이 38건으로 목표 30건을 초과 달성했으며,
기존 고객 재구매율은 68%로 전분기 대비 5%p 향상되었습니다.
그러나 영업비용이 8.5억원으로 예산 대비 15% 초과하여
순이익률은 전분기 21%에서 18%로 하락했습니다.
4분기 목표는 매출 50억원, 영업비용 8억원 이내 통제입니다.
"""

summary_prompt = f"""다음 보고서를 경영진을 위한 3줄 핵심 요약으로 만들어줘:\n{report_text}"""

print("=" * 50)
print("📊 Claude가 만든 경영진 요약 (3줄):")
print("=" * 50)
print(ask_ai(summary_prompt, model_type="claude"))

print("\n" + "=" * 50)
print("📊 GPT가 만든 경영진 요약 (3줄):")
print("=" * 50)
print(ask_ai(summary_prompt, model_type="gpt"))
```

## 🌡️ Temperature 파라미터 실험

**Temperature**는 AI 출력의 창의성/무작위성을 조절합니다:
- `0.0` : 항상 같은 답변 (업무용 권장)
- `1.0` : 매번 다른, 더 창의적인 답변 (창작용)

```python
# Temperature 비교 실험
test_prompt = "신제품 마케팅 슬로건을 한 줄로 만들어줘. 제품: 스마트 물병 (물 마실 시간을 알려주는 기능)"

# Claude Temperature 비교
print("✨ Claude (model=claude-3-5-sonnet-20240620)")
print("🔵 Temperature = 0.0 (일관된 답변)")
for i in range(3):
    response = claude_client.messages.create(
        model="claude-3-5-sonnet-20240620",
        max_tokens=100,
        temperature=0.0,   # 결정적: 항상 같은 답변
        messages=[{"role": "user", "content": test_prompt}]
    )
    print(f"  시도 {i+1}: {response.content[0].text.strip()}")

print()
print("🔴 Temperature = 1.0 (다양한 답변)")
for i in range(3):
    response = claude_client.messages.create(
        model="claude-3-5-sonnet-20240620",
        max_tokens=100,
        temperature=1.0,   # 창의적: 매번 다른 답변
        messages=[{"role": "user", "content": test_prompt}]
    )
    print(f"  시도 {i+1}: {response.content[0].text.strip()}")

print("\n" + "="*50 + "\n")

# GPT Temperature 비교
print("✨ GPT (model=gpt-4o)")
print("🔵 Temperature = 0.0 (일관된 답변)")
for i in range(3):
    response = gpt_client.chat.completions.create(
        model="gpt-4o",
        max_tokens=100,
        temperature=0.0,   # 결정적: 항상 같은 답변
        messages=[{"role": "user", "content": test_prompt}]
    )
    print(f"  시도 {i+1}: {response.choices[0].message.content.strip()}")

print()
print("🔴 Temperature = 1.0 (다양한 답변)")
for i in range(3):
    response = gpt_client.chat.completions.create(
        model="gpt-4o",
        max_tokens=100,
        temperature=1.0,   # 창의적: 매번 다른 답변
        messages=[{"role": "user", "content": test_prompt}]
    )
    print(f"  시도 {i+1}: {response.choices[0].message.content.strip()}")
```

## 📝 정리

| 개념 | 설명 |
|------|------|
| `client.messages.create()` | Claude에게 메시지를 보내는 핵심 함수 |
| `model` | 사용할 Claude 모델 (sonnet = 균형, haiku = 빠름, opus = 최강) |
| `max_tokens` | 응답 최대 길이 (1토큰 ≈ 0.75단어) |
| `temperature` | 창의성 조절 (0=일관성, 1=다양성) |
| `response.content[0].text` | 응답 텍스트 추출 방법 |
